### Import Necessary Libraries

```python
from pandas_datareader import data, wb
import pandas as pd
import numpy as np
import datetime
%matplotlib inline
```

## Create date Variables

```python
start= datetime.datetime(2006,1,1)
end=   datetime.datetime(2016,1,1)
```

```python
###  Get Remote Data
## Bank Of America  BAC
BAC = data.DataReader('BAC','yahoo',start,end)
## CitiGroup  C
C = data.DataReader('C','yahoo',start,end)
## Goldman Sachs GS
GS = data.DataReader('GS','yahoo',start,end)
##JPMorgan Chase  JPM
JPM = data.DataReader('JPM','yahoo',start,end)
##Morgan Stanley  MS
MS = data.DataReader('MS','yahoo',start,end)
## Wells Fargo  WFC
WFC = data.DataReader('WFC','yahoo',start,end)
```

### CREATE A LIST OF STOCK TICKER SYMBOLS (AS STRINGS) IN ALPHABETICAL ORDER, AND CALL THE LIST TICKERS 

```python
tickers = ['BAC','C','GS','JPM','MS','WFC']
```

### Use pd.cocat to concatenate the bank dataframes into a single dataframe called bank_stocks

```python
bank_stocks = pd.concat([BAC,C,GS,JPM,MS,WFC], axis=1, keys=tickers)
```

### Check the head of the bank_stocks dataframe

```python
bank_stocks.columns.names =(['ticker','stock info'])
```

```python
bank_stocks.head()
```

## EDA Exporatory Data Analysis

### Max Stock Close price for each of the banks through the period

```python
### Using for loop
for tick in tickers:
    print(tick,bank_stocks[tick]['Close'].max())
```

```python
### Using Pandas cross section xs()
bank_stocks.xs(key='Close', axis=1, level='stock info').max()
```

```python
### Create a new empty Returns dataframe
```
